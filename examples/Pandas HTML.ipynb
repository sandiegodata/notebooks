```python
bp = pd.read_html('http://blackdemographics.com/population/black-city-population/', header = 0)[0]
```

```python
bp['lcity'] = bp.iloc[:,1].apply(lambda x: x.split(',')[0].lower())
c = list(df.columns)
c[4] = 'Homicide'
df.columns = c
df['lcity'] = df.City.apply(lambda x: x.split(',')[0].lower())
bp2 = bp.set_index('lcity')
bp2['Percent Black'] = bp2['Percent Black'].apply(lambda x: float(x.strip('%'))/100)
df2 = df.set_index('lcity')
df3 = df2.join(bp2)
df3 = df3[['Violent Crime','Property Crime','Rape','Homicide','Percent Black']]
df3.dropna()
print
```

```python
pd.scatter_matrix(df3, alpha=0.2, figsize=(20,20))
print
```
