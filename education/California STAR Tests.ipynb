# California STAR Test Basic Group Analysis

This notebook presents basic analysis of the California STAR test data files, using the 2012 research file. Each figure below shows the density curves for a subgroup for each of a set of tests and grades, all taken from the 2012 STAR test score research files from the California Department of Education. 

The grades are: 
* 5
* 8
* 10

The tests, and their codes,  are: 
* 7: CST English-Language Arts
* 9: CST Algebra I
* 32: CST Science - Grade 5, Grade 8, and Grade 10 Life Science
 
See [The STAR website](http://star.cde.ca.gov/star2012/research_fixfileformat.aspx) for a description of all of the field codes.

## What's a Density Curve?

A Density curve is a lot like a histogram, but instead of showing you the number of students with scores in each band, it shows you the probability of finding a student with a particular score. However, for most use, and for visual analysis, the density curves can be used like histograms. They have approximately the same shape, but are smoother. 

The curves are generated with a Gaussian KDE using Scotts rule for bandwidth estimation. The data points are drawn from all schools in California.

## Notes

There are a few interesting features in these curves:
* For Grade 8, Test 32 (Science), the curves appear wider that for Grade 5 and Grade 10
* Grade 8 also appers wider for Test 9, Algebra.


```python
%matplotlib inline
import ambry
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

l = ambry.ilibrary()
p = l.get('cde.ca.gov-star-orig-star-2012').partition
```

```python
def plot_groups(df, ax, subgroups, test_id, grade, title):
    from scipy.stats import gaussian_kde
  
    ax.grid(color='w', linewidth=2, linestyle='solid')

    colors = 'k','r','b','g','m','c'

    for i, subgroup in enumerate(subgroups):
        data = df[(df['subgroup'] == subgroup) & 
                  (df['test_id'] == int(test_id)) &  
                  (df['grade'] == grade)].mean_score.dropna()
     
        ax.set_title(title)
        
        if len(data.index) < 2:
            continue
            
        #print subgroup, test_id, grade, len(data)
        
        density = gaussian_kde(data,'scott')
        xs = np.linspace(0,data.max(),200)
        ax.plot(xs,density(xs), color=colors[i], label=str(subgroup))
        ax.set_xlim(200, 600)
        ax.set_ylim(0,0.025)
        
def get_data(subgroups):
    df = p.select("""
    SELECT subgroup, grade, test_id, mean_score from star 
    WHERE subgroup IN ({subgroups}) AND school != 0"""
                  .format(subgroups=','.join([str(x) for x in subgroups]))).pandas

    return df
 
def plot_groups_grid(df, subgroups, grades, tests, swap=False):
    
    if swap:
        nrows, ncols = len(tests),len(grades)
    else:
        nrows, ncols = len(grades),len(tests)
        
    
    fig, axes = plt.subplots(nrows = nrows, ncols=ncols, 
                         #sharex='col', 
                         sharey='row', figsize=(15, 15), 
                         subplot_kw={'axisbg':'#EEEEEE', 'axisbelow':True})

   
    for col, test_id in enumerate(tests):
        for row, grade in enumerate(grades):
            plot_groups(df, axes[row][col] if not swap else axes[col][row],
                        subgroups, test_id, grade, 'test={} grade={}'.format( test_id, grade))

    plt.legend( ncol = len(subgroups), mode="expand", loc = 'lower center', bbox_to_anchor = (0,.05,1,1),
                bbox_transform = plt.gcf().transFigure, borderaxespad=0., prop={'size':18} )
    

            
    
```

## Parent Education

Density curves by parental education:
* 90 (Black) Not a High School Graduate	
* 91 (Red) High School Graduate
* 92 (Blue) Some College (Includes AA Degree)
* 93 (Green) College Graduate
* 94 (Magenta) Graduate School/Postgraduate Training
* None Parent Education -- Declined to State

```python
subgroups = [90,91,92,93,94]
df = get_data(subgroups)
tests = [7,9,32]
grades = [5,8,10]

plot_groups_grid(df, subgroups, grades, tests)
```
