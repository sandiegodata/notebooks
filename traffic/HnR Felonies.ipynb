```python
%matplotlib inline
import pandas as pd
df = pd.read_csv('http://devtest.sandiegodata.org.s3.amazonaws.com/manifests/7a007ac7-960f-468b-8ce3-7b9f83e34ea8/hnr0914.csv')
```

```python
df[['incar','prob','work', 'prog']].fillna(0).hist(bins=50,figsize=(15,10))
```

This shows that most sentences are relatively short, across all sentences. Only probation has many long sentences, although there are 2 incarcerations at more than 3,000 days.  

Next, we create a dataframe that has the means and counts for each of the major categories of sentences, Incarceration, Probation and Work programs, grouped by the `highest charge`. This frame excludes `highest_charge` values where there is only one record with that charge. 

```python
scol = ['highest_charge',u'drug', u'drug_patient', u'drug_res', u'full', u'incar', u'incar_county', 
        u'incar_life', u'incar_prison', u'incr_local', u'prob', u'prob_formal',
        u'prob_sum', u'prog',  u'prog_child', u'prog_dv', u'work', u'work_furl',
        u'work_pub', u'work_vol']

dfg = df[['highest_charge','incar','prob','work']].fillna(0).groupby('highest_charge')
means = dfg.mean()
means['counts'] = dfg.count()['highest_charge']
means['incar_total'] = means.incar * means.counts
means['prob_total'] = means.prob * means.counts
means['work_total'] = means.work * means.counts

means.to_csv('fhnr_highest_charge.csv')
```

Download [fhnr_highest_charge.csv](fhnr_highest_charge.csv)

Here are the mean values, sorted by the `incar_total` column. The `..._total` columns multiply the average for that sentence type with the number of sentences of that type. It's the total amount of sentence time for each `highest_charge`, and estimates the significance of that charge in terms of person-days of punishment. The highest values charge for both incarceration and probation, [VC23153(a) charge](https://www.dmv.ca.gov/pubs/vctop/d11/vc23153.htm), is:

    It is unlawful for a person, while under the influence of any alcoholic beverage to drive a 
    vehicle and concurrently do  any act forbidden by law, or neglect any duty imposed by law in driving 
    the vehicle, which act or neglect proximately causes bodily injury to any person other than the driver.
    
The second, [VC20001(a)](https://www.dmv.ca.gov/pubs/vctop/d10/vc20001.htm), is:

    (a) The driver of a vehicle involved in an accident resulting in injury to a person, other than himself or herself, 
    or in the death of a person shall immediately stop the vehicle at the scene of the accident and shall fulfill 
    the requirements of Sections 20003 and 20004.
    
PC191 is Manslaughter, and PC187 is Murder. ( "VC" means Vehicle Code, and "PC" means Penal Code. ) Note that there are two 187 entries and two 191 Entries ( PC191.5(a) and PC191.5(b), etc. ) 

So, the most common `highest_charges` both in terms of number of counts and person-days of punishment are drunk driving resulting in injury, while the longest sentences are for drunk driving resulting in death.
