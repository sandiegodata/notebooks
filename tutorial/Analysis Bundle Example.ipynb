```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ambry

ab = ambry.new_analysis_bundle(source='sandiegodata.org', dataset='crime',
                        subset='test', bspace='casnd', revision=1)
ab
```

Set some configuration for the bundle. 

```python
with ab.session as s:
    ab.partitions.clean(s)

a = ab.metadata
a.title = 'An Analysis Bundle Example Using Crime Data'
a.tags = ['example','another']
a.groups = ['examples']

ab.metadata.dependencies =  {
    'incidents' : 'clarinova.com-crime-incidents-casnd-linked-crimes',
    'placestats' : 'clarinova.com-places-casnd-areastats',
    'addresses' : 'clarinova.com-geocode-casnd-address',
    'roads' : 'sangis.org-streets-orig-roads-geo' 
}

# Every partition needs to have a table, so to make a temporary partition to attach other to
# we need to create a temp table. 
with ab.session:
    t = ab.schema.add_table('temp')
    ab.schema.add_column(t,'id', datatype='integer')
```

```python

dbp = ab.partitions.new_geo_partition(table='temp', format='geo')

incidents = ab.library.dep('incidents').partition
placestats = ab.library.dep('placestats').partition

incidents.attach(placestats,'placestats')
```

```python
incidents.table
```

```python
q = """
SELECT crimes.*, 
city_stats.pop as city_pop, city_stats.land as city_area, 
CAST(city_stats.pop AS REAL)/ CAST(city_stats.land AS REAL) as city_density,
community_stats.pop as community_pop, community_stats.land as community_area, community_stats.name as community_name, 
CAST(community_stats.pop AS REAL)/ CAST(community_stats.land AS REAL)*1000000 as community_density 
FROM crimes 
LEFT JOIN placestats.areastats AS city_stats ON city_stats.type = 'city' AND city_stats.code = crimes.city
LEFT JOIN placestats.areastats AS community_stats ON community_stats.type = 'community' AND community_stats.code = crimes.community
WHERE community != '-'
LIMIT 50000
"""

inct = incidents.select(q).pandas
inct.count()
```

```python
p = ab.partitions.new_db_from_pandas(inct,table = 'by_community')
```

Now we have a new partition in the bundle. Let's verify that the partition exits and can be used normally. 

```python
inct = ab.partitions.find(table = 'by_community').select("SELECT * FROM by_community").pandas
dfm= inct.groupby(['hour','legend'])
counts = dfm.count()
cus = counts.unstack(1)
uscounts = cus['id']
ax = uscounts.plot(subplots=True, figsize=(10,20), colormap='jet')
```
