Crime Analysis Examples
=======================

This notebook demonstrates how to use Ambry and Pandas for exploratory crime analysis. The analysis will examine six 
years of data that was aggregated by SANDAG and processed by the [San Diego Regional Data Library](http://sandiegodata.org). The data is published in CSV and shapefile form at the [Library's data repository](http://data.sandiegodata.org/dataset/clarinova_com-crime-incidents-casnd-7ba4-extract), but this analysis uses an updated dataset that is only available through Ambry. 

Ambry is our data package management system, which is currently in alpha testing, with an expected public release in Spring of 2014. 


## Initializing the Data

The first step is to get the data. We'll use the Ambry library to get two data files, link them together and run a query, which will place data into a Pandas DataFrame.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ambry

l  =  ambry.ilibrary()

incidents = l.get('clarinova.com-crime-incidents-casnd-linked-crimes').partition
placestats = l.get('clarinova.com-places-casnd-areastats').partition
```

In Ambry, we can use the library to find datasets and inspect their columns. For instance, here are the fields in the`incidents` table. The code `incidents.table` would give you the schema for the tables, but we'll omit the output to leave space for the analysis outputs. 

In Ambry, datasets are Sqlite files, so we can connect them together and run queries across them.  


```python
incidents.attach(placestats, 'placestats')
```

Now, the `incidents` database has an attached database for place statistics, so we can get population quantities and densities and full community names. The next step is to query the combined datasets and do some processing to get specific names for areas.  After defining the query, you can see how simple it is to get the data into a Pandas DataFrame. 

The Placestats data base has an `areastats` table that has records for neighborhoords, communities, cites and council districts in San Diego county, although the communities and neighborhods are only specified for the city of San Diego, So, the `WHERE community != '-'` means that all of these records will be for San Diego city, which is about half of the crime records for the county. 

The `LEFT JOIN` clauses attach parts of the `areastats` table to codes in the crime dataset. The library uses 6 charater codes for each of the cities and neighborhoods. These codes are memorable once you learn them -- for instance, the codes for "Pacific Beach" is "SanPCF" -- but they are difficult to remember when you are dealing with a lot of them. 


```python
q = """
SELECT crimes.*, 
    city_stats.pop as city_pop, city_stats.land as city_area, 
    CAST(city_stats.pop AS REAL)/ CAST(city_stats.land AS REAL) as city_density,
    community_stats.pop as community_pop, community_stats.land as community_area, community_stats.name as community_name, 
    CAST(community_stats.pop AS REAL)/ CAST(community_stats.land AS REAL)*1000000 as community_density 
FROM crimes
LEFT JOIN placestats.areastats AS city_stats ON city_stats.type = 'city' AND city_stats.code = crimes.city
LEFT JOIN placestats.areastats AS community_stats ON community_stats.type = 'community' AND community_stats.code = crimes.community
WHERE community != '-'
"""

inct = incidents.select(q).pandas
inct.head(5)
```

```python
inct['id'].count()
```

Now we've got 400,000 records for crime incidents in San Diego, and we can start to visualize them. 

# Time Series Visualizations

The first visualizations will look at all crime incidents over the entire period by hour of day and then day of week, as line charts. Then we'll look at a different way to view the same data.  First, here are the number of crimes, by category, by hour.

## Hour and Day of Week Heatmaps

In the previous charts, the Y axis scale isn't very important -- you probably were more interested in the relative changes in the number of crimes than in the absolute number. By dispensing with the numeric asix and replacing it with a color, we can combine both charts and make them more visually appealing, with a heatmap. 

Here is the function to create the grid of heatmaps. 

```python
from math import ceil

def plot_heat_grid(df, legends = None):

    if not legends:
        legends = sorted([ l for l in df.legend.unique() if l.strip() and l not in ('-', 'ARSON','HOMICIDE') ])

    c = len(legends)
    ncols = 4.
    nrows = ceil(float(c)/ncols)

    cmap = plt.get_cmap('YlOrRd')

    fig, axes = plt.subplots(nrows=int(nrows), ncols=int(ncols), figsize=figsize(10,10), squeeze=True, sharex=False, sharey=False)

    fig.set_size_inches(ncols*2.5,nrows*2.5)
    axes = axes.ravel() # Convert 2D array to 1D
    plt.tight_layout(h_pad=1)
    for i, legend in enumerate(legends):

        sub = df[df.legend == legend]
        heatg = sub.groupby(['dow','hour'])
        hgcounts = heatg.count()['id'].unstack('hour').fillna(0)
   
        # Converting to an array puts it into a for that
        # matplotlib expects. This probably only works b/c the
        # hours and days of week are 1-based indexes. 
        axes[i].pcolormesh(array(hgcounts.T),cmap=cmap)
        axes[i].set_title(legend.title())
    
```

For the first grid, we'll look at crime types, across all of San Diego. The vertical axis is hour of day, and the horizontal axis is day of week, with Sunday being the area between 0 and 1, etc.  The color indicates the number of crimes in that hour of the week, with red being larger than yellow. Note that each grid is scaled independently, so a red square in one heatmap is not comparable to a red square in another heatmap. This visualization is only for looking at the paterns of crimes over time. 

```python
plot_heat_grid(inct)
```

There are a lot of stories in this grid!

* The "12:00 default" is clearly visible in the Fraud and Theft heatmaps. 
* Burglary, usually theft from a person's home, happens during the times that people are either at work, or out on a Friday evening. 
* DUIs happen around when the bars close. 
* Prostitution is most popular on Thursday evenings. ( Or that's when the vice squad likes to hit the town. )

## Comparing Communities

The preceeding analysis lets us see which of the crime types have interesting patterns. Now let's look at how those patterns vary across communities. There there are a lot of communities, we'll pick a selection of them. The selection includes the largest communities a few of the poorest and a few of the wealthy ones. 

```python
# Get a list of all of the communities, ordered by population, so we can manually extract codes. 
x = inct[['legend','community']]
#x = x[x.legend == 'SEX CRIMES']
x.groupby('community').count().sort('legend', ascending=False)


communities = ['SanDOW','SanMDC','SanPCF','SanSOT','SanGRE','SanCLA','SanUPT','SanENC','SanUNV','SanMID',
               'SanMIR','SanSKY','SanOAT', 'SanCOL', 'SanLAJ']
```

Here is a variant of the heat grid function, altered to display communities. 

```python
def plot_heat_grid_community(df,  communities=None):

    if not communities:
        communities = sorted([ l for l in df.community.unique() if l.strip() and l not in ('-') ])

    c = len(communities)
    ncols = 5.
    nrows = ceil(float(c)/ncols)

    cmap = plt.get_cmap('YlOrRd')

    fig, axes = plt.subplots(nrows=int(nrows), ncols=int(ncols), figsize=figsize(10,10), squeeze=True, sharex=False, sharey=False)

    fig.set_size_inches(ncols*2,nrows*2)
    axes = axes.ravel() # Convert 2D array to 1D
    plt.tight_layout(h_pad=1)
    for i, community in enumerate(communities):

        sub = df[df.community == community]
        if not sub.empty:
            heatg = sub.groupby(['dow','hour'])

            hgcounts = heatg.count()['id'].unstack('hour').fillna(0)

            # Converting to an array puts it into a for that
            # matplotlib expects. This probably only works b/c the
            # hours and days of week are 1-based indexes. 
            axes[i].pcolormesh(array(hgcounts.T),cmap=cmap)
            axes[i].set_title(community)
        
```

```python
inct.legend.unique()
```

First up, Sex crimes, emphasizing prostitution. Most of the grids are spotty, indicating there are not enough incidents for proper analysis, with the exception of North Park, Downtown and Mid-City / City Heights. 

```python
plot_heat_grid_community( inct[ ( inct.legend == 'SEX CRIMES') & (inct.description == 'PROSTITUTION' ) ], communities=communities)
```

Burglary shows some interesting patterns -- several of the communities are missing the cluster of incidents in the middle of the workday, and some have much stronger Friday evenings. Our original analysis showed the workday and Friday evening patterns on the same heatmap, but it turns out they are independent, and occur in different communities. 

A guess at the reason is that in the more affluent areas there are more people who work from home or maintain the home, so the burglars have to pick another time. A future analysis could link in Census data to verify this hypothesis. 


```python
plot_heat_grid_community( inct[inct.legend == 'BURGLARY'], communities=communities)
```

DUIs however, look pretty much the same everywhere, although some communities get started earlier. The communities that have more "noise" in the middle of the map, rather than being solid yellow, are ones that have fewer DUIs; in the comunities with a lot of DUIs, such as SanDOW and SanPCF the large number of nighttime DUIs swamps those during the day. 

```python
plot_heat_grid_community( inct[inct.legend == 'DUI'], communities=communities)
```

The patterns for Theft are more subtle, but seem to be connected to whether the community has an active nightlife, where more of the interactions between people occur in the evening. 

```python
plot_heat_grid_community( inct[inct.legend == 'THEFT/LARCENY'], communities=communities)
```
